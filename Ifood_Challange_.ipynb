Suppose we are trying to attract more customers. In our previous data analysis, we saw that people without children spend more than people with more children.

Our marketing team thought about making campaigns aimed at the group of people with children (regardless of the number), but we need to verify if in fact the average spending of the groups is different.

Therefore:

* Group all people who have children into a group called "with children" and compare the average spending of people without children and people with children.
Afterwards, compare the groups of people with children.
* The marketing team asked us to check if there are any differences in expenses (expenses) in those who have 1, 2, 3 or more children so they can further narrow the campaign.
* Is the mean of these groups different? Which one(s) has the lowest average?

# Comparing the amount spend by people without children to people with children

```python
df['children_status'] = df['kids'].apply(lambda x: 'with children' if x > 0  else 'without children')
```

```python
df.groupby('children_status')['MntTotal'].mean()
```

```python
fig = px.box(df, x="children_status", y="MntTotal")
fig.update_traces(marker_color="#506369")
fig.update_layout(title_text='Boxplot expenses with and without kids',height=700,width=1300, plot_bgcolor='#e8D3D0',
    font_family="Courier New",
    font_color="#506369",
    title_font_family="Times New Roman",
    title_font_color="#506369",
    legend_title_font_color="#506369"
)
fig.update_yaxes(showgrid=False,zeroline=False, visible=False,hoverformat=False)
fig.update_xaxes(tickmode='linear',showgrid=False,zeroline=False, visible=True, hoverformat=False,categoryorder='total descending')
fig.show()
```

Seeing there is some overleap between the box plots make us unsure about the difference between them.

```python
df.groupby('children_status')['MntTotal'].count()
```

Here we have a comparision of the mean between two groups:
* The category is independent (have children or not);
* The variable is continuous (amount spent);
* We have a good amount of data in each sample;
* Populational standant deviation is unknown.

So, we choose to use a **T test**.

Our hypotesis are:

* H0: the means are equal (u1=u2)
* H1: the means are not equal (u1!=u2)


```python
with_children_df=df.query("children_status == 'with children'")
```

```python
with_children=with_children_df['MntTotal']
```

```python
with_children
```

```python
without_children=df.query("children_status == 'without children'")
```

```python
without_children=without_children['MntTotal']
```

```python
with_children
```

```python
stats.ttest_ind(a = without_children,b = with_children,equal_var=True)
```

The p value is so low that we can reject the null hypothesis.
Which means the mean value spent by people without kids is higher than the one spent by people with kids.

Now, let's see if there's a difference between the number of kids in each family.

```python
with_children_df.head()
```

We intend to use 1 way ANOVA to verify if there's somw difference between the means. But first, there are some things we have to check:

* The samples are random.
* The groups are independent.
* The independent variable is category.
* The data follows a normal distribution or we have more than 15 observations.
* The variance in each group are very similar.


```python
fig = px.box(with_children_df, x="kids", y="expenses")
fig.update_traces(marker_color="#506369")
fig.update_layout(title_text='Boxplot expenses per number of kids',height=700,width=1300, plot_bgcolor='#e8D3D0',
    font_family="Courier New",
    font_color="#506369",
    title_font_family="Times New Roman",
    title_font_color="#506369",
    legend_title_font_color="#506369"
)
fig.update_yaxes(showgrid=False,zeroline=False, visible=False,hoverformat=False)
fig.update_xaxes(tickmode='linear',showgrid=False,zeroline=False, visible=True, hoverformat=False,categoryorder='total descending')
fig.show()
```

Seeing there is some overleap between the box plots make us unsure about the difference between them.

We are sure about all the hypothesis, except the one about the variance, so let's test it.

We'll perform a Bartlett test:
* The null hypothesis is that all the variances are equal.
* The alternative hypothesis is that one or more of the variance are different. 


```python
list_1c = with_children_df[with_children_df["kids"] == 1]["expenses"].dropna()
list_2c = with_children_df[with_children_df["kids"] == 2]["expenses"].dropna()
list_3c = with_children_df[with_children_df["kids"] == 3]["expenses"].dropna()
```

```python
stats.bartlett(list_1c, list_2c, list_3c)
```

The pvalue is really low, so we can assume the variances are not the same. In this case, let's use the Welch ANOVA.

```python
pg.welch_anova(dv='expenses', between='kids', data=with_children_df)
```

As the pvalue is very low we can reject the null hypothesis and conclude all the mans are not the same. Now, let's see which ones are difference. We are going to use the Games Howell test to see the differences. The Reason we chose this test is because it doesn't assume normality, same variance and size of the samples to be the same.

```python
pg.pairwise_gameshowell(dv='expenses', between='kids', data=with_children_df)
```

Pvalue in the groups who have 1 and 2 kids are very low, so we can assume the means are different.

Pvalue in the groups who have 1 and 3 kids is also low, so we can assume their means are also different.

But Pvalue in groups 2 and 3 is not low, so in these groups we can assume the means are equal.
